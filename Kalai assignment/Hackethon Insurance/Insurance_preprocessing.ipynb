```python
X=[x for x in dataf.columns if x not in (['ID','source','TARGET'])]
X_train=dataf[X]
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
df_mod = sc_X.fit_transform(X_train)

df_mod=pd.DataFrame(df_mod,columns=X)
df_mod=pd.concat([df_mod,dataf[['ID','source']]],axis=1)

```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
n_components=20
pca = PCA(n_components)
cols=[]
for i in range(n_components):
    nm='PCA'+str(i+1)
    cols.append(nm)
```

```python
X=[x for x in df_mod.columns if x not in (['ID','source','TARGET'])]


principalComponents = pca.fit_transform(df_mod[X])

PCAdf = pd.DataFrame(data = principalComponents
             , columns = cols)
```

```python
dataf['target_value']= np.where (dataf.TARGET > 0,1,0)
```

```python


PCAdf=pd.concat([PCAdf,dataf['target_value']],axis=1)
```

```python

PCAdf=pd.concat([PCAdf,dataf[['ID','source']]],axis=1)
```

```python
train_modified=PCAdf.loc[PCAdf.source=='train']
test_modified=PCAdf.loc[PCAdf.source=='test']
```

```python
test_modified.drop(columns=['source'],inplace=True)
train_modified.drop(columns=['source'],inplace=True)
```

```python
train_modified.to_csv("train_ready_for_model_loan.csv",index=False)
test_modified.to_csv("test_ready_for_model_loan.csv",index=False)
```
